## 1) Three-way merge sort
*#sort Implement three-way merge sort in Python. It should at a minimum accept lists off integers as input.*

```python
from math import ceil

def three_way_merge_sort(a):
    if len(a)<2:
        return a #checks if already sorted
    
    m= ceil(len(a)/3) # division point
                        
    
    b=[] # empty list to store sorted lists for merging
    for i in range(0,len(a),m): # iterate through a in segments of size m
        b.append(three_way_merge_sort(a[i:i+m])) # recursively calls the merge sort function
     
    c=[] # making a new list to hold the merged list
    while len(b)>0: # we need to empty each of the lists in b into c
        minimum=b[0][0] # initializing the minimum (first of first list)
        index=0 # setting the index to 0
        for i in b:  # for each of the lists in b
            if i[0]<minimum: # check if new minimum is needed
                minimum=i[0] # change minimum
                index=b.index(i) # set index to position of list
        c.append(minimum) # append the minimum value to c
        b[index].pop(0) # remove minimum value from b
        if len(b[index])==0: # remove empty lists
            b.remove(b[index])
    return c

x=[32,4,13,15,1,2,35,1,2,5,1,23]
three_way_merge_sort(x)


#comment: I made the k-way merge sort first and this is just an adaptation of it
```

## 2) Insertion-merge-sort
*#sort Implement a second version of three-way merge sort that calls insertion sort when sublists are below a certain length (of your choice) rather than continuing the subdivision process.*

```python
 from math import ceil

def insertion_sort(a):
    for j in range(1,len(a)): # for each element in the list
        key= a[j] # define a key that is equal to the element in position j
        i = j-1 # and an i which is the index of the element before the one in position j
        while i>=0 and a[i]>key: # checks if insertion should happen
            a[i+1]=a[i] # insert the new value into the correct position
            i-=1
        a[i+1]=key #move on to the next step
    return a #return the sorted list

def three_way_merge_insertion_sort(a):
    if len(a)<2:
        return a #checks if a is smaller than two elements, in which case it is already sorted.
    if len(a)<6:
        insertion_sort(a) # for lists shorter than 6 elements, use insertionsort instead.
    else:
        m= ceil(len(a)/3) # determine break points
    
        b=[] # empty list to store sorted lists for merging
        for i in range(0,len(a),m): # iterate through a in segments of size m
            b.append(three_way_merge_sort(a[i:i+m])) # recursively calls the merge sort function
     
        c=[] # making a new list to hold the merged list
        while len(b)>0: # we need to empty each of the lists in b into c
            minimum=b[0][0] # initializing the minimum (first of first list)
            index=0 # setting the index to 0
            for i in b:  # for each of the lists in b
                if i[0]<minimum: # check if new minimum is needed
                    minimum=i[0] # change minimum
                    index=b.index(i) # set index to position of list
            c.append(minimum) # append the minimum value to c
            b[index].pop(0) # remove minimum value from b
            if len(b[index])==0: # remove empty lists
                b.remove(b[index])
        return c

x=[32,4,13,15,1,2,35,1,2,5,1,23]
three_way_merge_insertion_sort(x)
```

**Figure 2:** A table comparing the complexity of the three algorithms.<br><br>
Complexity: All of the three algorithms above are recursive algorithms that use a divide-and-conquer approach. Their complexity will depend on their depth as well as the amount of work that needs to be done at each level (and at the very last level).<br>
For the three-way merge sort, there are n/3 levels and at each level there are 3 lists. The time-complexity of the algorithm therefore is <br>
$T(n)=3T(n/3)+O(n) -> O(n log3(n))$ <br>
Similarily, the time complexity of a basic merge sort is<br>
$T(n)=2T(n/2)+O(n) -> O(nlog2(n))$<br>
This would suggest that the three-way merge sort is better than the two-way merge sort, however it does not consider the time it takes for comparisons to be executed. Pracically, the three-way merge sort should only be better for short lists and the two-way merge sort is generally preferred.<br>
The three_way_insertion_sort algortithm reduces the depth of the algorithm. After that level, it follows the complexity of insertion sort (which O(n^2)). At first, this seems slower, but for very short lists, insertion sort actually performs better than merge sort.<br>
To find the complexity, we have to add the new complexity of the merge sort at it's new depth to the complexity of the insertion sort. To visualize this better, refer to the graph below.<br>
New three-way-merge-sort complexity: It still needs O(n) on each level to merge them back together, but we are effectively dividing n into groups of 6, so $T(n)= 3T(n/6/3)+O(n)$ . Therefore, the new complexity of the merge sort part is $O(nlog18(n))$. To this we add the complexity of the insertion sort. It will have n/6 lists to sort and a complexity of O(n) in the best case and O(n^2) in the worst case. for lists of length 6 this is 6^2 operations. The total complexity is therefore $O(6n +nlog18(n))$ <br> <br>
The insertion sort can augment merge sort because it has a lower running time for lower values of n. The results of my experiments confirm this hypothesis.<br><br>
